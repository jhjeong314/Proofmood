```python
# Distributivity of 'and' over 'or' (left)
# A and (B or C) ├─ (A and B) or (A and C)
prf_str = '''
1. A and (B or C) .hyp
proves
2. A 
3. B or C 
  4. B .hyp
  proves
  5. A and B 
  6. (A and B) or (A and C) 
  7. C .hyp
  proves
  8. A and C 
  9. (A and B) or (A and C) 
10. (A and B) or (A and C) 
'''
proof = ProofNodeS(parse_fitch(prf_str))
proof.search_proof(1)
proof.show_fitch_text()
```

```python
# Distributivity of 'and' over 'or' (right)
# (A and B) or (A and C) ├─ A and (B or C)
prf_str = '''
1. (A and B) or (A and C) .hyp
proves
  2. A and B .hyp
  proves
  3. A 
  4. B 
  5. B or C 
  6. A and (B or C) 
  7. A and C .hyp
  proves
  8. A 
  9. C 
  10. B or C 
  11. A and (B or C) 
12. A and (B or C) 
'''
proof = ProofNodeS(parse_fitch(prf_str))
proof.search_proof(1)
proof.show_fitch_text()
```

```python
# Distributivity of 'or' over 'and' (right)
# (A or B) and (A or C) ├─ A or (B and C)
prf_str = '''
1. (A or B) and (A or C) .hyp
proves
2. A or B 
3. A or C 
  4. A .hyp
  proves
  5. A or (B and C) 
  6. B .hyp
  proves
    7. C .hyp
    proves
    8. B and C 
    9. A or (B and C) 
    10. A .hyp
    proves
    11. A or (B and C) 
  12. A or (B and C) 
13. A or (B and C) 
'''
proof = ProofNodeS(parse_fitch(prf_str))
proof.search_proof(1)
proof.show_fitch_text()
```

```python
# A imp B, not B ├─ not A
# modus tollens

prf_str = '''
1. A imp B .hyp
2. not B .hyp
proves
  3. A .hyp
  proves
  4. B 
  5. bot 
not A 
'''
proof = ProofNodeS(parse_fitch(prf_str))
proof.search_proof(1)
proof.show_fitch_text()
```

```python
# A or B, not A ├─ B
# modus tollendo ponens

prf_str = '''
1. A or B .hyp
2. not A .hyp
proves
  3. A .hyp
  proves
  4. bot 
  5. B 
  6. B .hyp
  proves
  7. B 
B 
'''
proof = ProofNodeS(parse_fitch(prf_str))
proof.search_proof(1)
proof.show_fitch_text()
```

```python
# not(A and B), A ├─ not B
# modus ponendo tollens

prf_str = '''
1. not(A and B) .hyp
2. A .hyp
proves
  3. B .hyp
  proves
  4. A and B 
  5. bot 
6. not B 
'''
proof = ProofNodeS(parse_fitch(prf_str))
proof.search_proof(1)
proof.show_fitch_text()
```
