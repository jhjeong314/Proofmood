```python
# Axiom 2 (Hilbert, Lukasiewicz)

prf_str = '''
1. .hyp
proves
  2. A imp (B imp C) .hyp
  proves
    3. A imp B .hyp
    proves
      4. A .hyp
      proves
      5. B
      6. B imp C 
      7. C
    8. A imp C
  9. (A imp B) imp (A imp C)
10. (A imp (B imp C)) imp ((A imp B) imp (A imp C))
'''
proof = ProofNodeS(parse_fitch(prf_str))
assert isinstance(proof, ProofNodeS)
proof.search_proof(1)
proof.show_fitch_text()
```

```python
# Transitivity of Implication

prf_str = '''
1. .hyp
proves
  2. A imp B .hyp
  proves
    3. B imp C .hyp
    proves
      4. A .hyp
      proves
      5. B 
      6. C 
    7. A imp C 
  8. (B imp C) imp (A imp C) 
9. (A imp B) imp ((B imp C) imp (A imp C)) 
'''
proof = ProofNodeS(parse_fitch(prf_str))
assert isinstance(proof, ProofNodeS)
proof.search_proof(1)
proof.show_fitch_text()
```

```python
# Transitivity of implication may be easier to understand
# if we write it as follows:
#   (A imp B) and (B imp C) imp (A imp C)
# In fact, the following is a tautology.

prf_str = '''
1. .hyp
proves
  2. A imp (B imp C) .hyp
  proves
    3. A and B .hyp
    proves
    4. A 
    5. B imp C 
    6. B 
    7. C 
  8. A and B imp C 
  9. A and B imp C .hyp
  proves
    10. A .hyp
    proves
      11. B  .hyp
      proves
      12. A and B 
      13. C 
    14. B imp C 
  15. A imp (B imp C) 
16. (A and B imp C) iff (A imp (B imp C)) 
'''
proof = ProofNodeS(parse_fitch(prf_str))
assert isinstance(proof, ProofNodeS)
proof.search_proof(1)
proof.show_fitch_text()
```
